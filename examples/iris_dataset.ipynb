# Iris Dataset Visualization

```python
from pprint import pprint
from data_catalog.assets import AssetService
import os
from dotenv import load_dotenv
load_dotenv();
```

### Finding the asset

```python
# Reads DC_USERNAME and DC_PASSWORD from the environment variables (can be provided in .env file)
with AssetService(username=os.getenv('DC_USERNAME'), password=os.getenv('DC_PASSWORD')) as asset_service:
    asset = asset_service.get_asset('5ffc9f134732492bef18336f')
```

   ### Getting the data

```python
dataframe = asset.get_data()
dataframe.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

print(dataframe)
```

### Visualization

```python
import seaborn as sns

sns.pairplot(dataframe, hue='class');
```
